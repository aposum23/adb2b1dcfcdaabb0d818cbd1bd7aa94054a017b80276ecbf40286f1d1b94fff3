```python
from skimage.io import imread, imshow, imsave
import cv2 as cv
import matplotlib.pyplot as plt
from skimage import img_as_float
import skimage
import numpy as np
from os import listdir
import os
import json
import sys
```

```python
JSON_PATH = "C:/Users/HAXF13D/Desktop/addition_traning_data/all_json"
PICTURES_PATH = "C:/Users/HAXF13D/Desktop/addition_traning_data/all"
WRITE_PATH = "C:/Users/HAXF13D/Desktop/addition_traning_data/masks"
```

```python
json_files = listdir(JSON_PATH)
pictures = listdir(PICTURES_PATH)
```

```python
for js in json_files:

    print(f"{JSON_PATH}/{js}")

    f = open(f"{JSON_PATH}/{js}")
    data = json.load(f)

    back_ground = np.zeros((224, 224), dtype=np.uint8)
    teeths = np.zeros((224, 224), dtype=np.uint8)
    carieses = np.zeros((224, 224), dtype=np.uint8)

    mask = np.zeros((224, 224), dtype=np.uint8)

    mask_teeths = data.get('tooth')
    mask_carieses = data.get('cavity')

    if mask_teeths is None:
        mask_teeths = []

    if mask_carieses is None:
        mask_carieses = []


    for i in range(0,len(mask_teeths),4):
        y_start = mask_teeths[i]
        y_finish = mask_teeths[i + 2]
        x_start = mask_teeths[i + 1]
        x_finish = mask_teeths[i + 3]

        mask[x_start:x_finish, y_start:y_finish] = 1
        back_ground[x_start:x_finish, y_start:y_finish] = 1
        teeths[x_start:x_finish, y_start:y_finish] = 1


    for i in range(0,len(mask_carieses),4):
        y_start = mask_carieses[i]
        y_finish = mask_carieses[i + 2]
        x_start = mask_carieses[i + 1]
        x_finish = mask_carieses[i + 3]

        mask[x_start:x_finish, y_start:y_finish] = 2
        back_ground[x_start:x_finish, y_start:y_finish] = 1
        carieses[x_start:x_finish, y_start:y_finish] = 1


    back_ground[back_ground == 0] = 2
    back_ground[back_ground == 1] = 0
    back_ground[back_ground == 2] = 1
    
    img = imread(f"{PICTURES_PATH}/{js.split('.')[0]}.jpg")
    imshow(img)
    
    #back_ground = cv.resize(back_ground, (256, 256))
    #carieses = cv.resize(carieses, (256, 256))
    #teeths = cv.resize(teeths, (256, 256))
    #img = cv.resize(img, (256, 256))
    
    
    os.remove(f"{PICTURES_PATH}/{js.split('.')[0]}.jpg")
    
    imsave(f"{PICTURES_PATH}/{js.split('.')[0]}.jpg", img)
    
    imsave(f"{WRITE_PATH}/background/{js.split('.')[0]}.png", back_ground)
    imsave(f"{WRITE_PATH}/caries/{js.split('.')[0]}.png", carieses)
    imsave(f"{WRITE_PATH}/teeth/{js.split('.')[0]}.png", teeths)
    #imsave(f"{WRITE_PATH}/{js.split('.')[0]}.png", mask)
```
